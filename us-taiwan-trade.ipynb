```python
import os
import requests
from urllib.parse import urljoin

base_url = "https://www.trade.gov.tw"

file_tag = soup_doc.find('a', href=re.compile(r"^/App_Ashx/File.ashx\?FileID="))
file_href = file_tag['href']

full_url = urljoin(base_url, file_href)
print(f"Full file download link: {full_url}")

file_name = file_tag['title']

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36',
    'Accept': 'application/octet-stream',  # Indicates we're expecting a file
}

response = requests.get(full_url, headers=headers)

with open(file_name, "wb") as f:
    f.write(response.content)
    print(f"File saved as {file_name}")
```

```python
df = df.drop([0, 1, 2, df.index[-3], df.index[-2], df.index[-1]]).reset_index(drop=True)
df
```

```python
df = df.loc[:, ~df.columns.isna()]
df
```

```python
df = df[~df.iloc[:, 1:].isna().all(axis=1)]
df
```

```python
df.columns = ['Month/Year', 'Total', 'Export', 'Import', 'Balance']
df
```

```python
# Updated transform_date function to handle non-string values
def transform_date(value):
    # Convert value to string
    value = str(value)
    # Handle 年 (year conversion)
    if '年' in value:
        year = int(value.replace('年', '').strip())
        return 1911 + year  # Convert from 民國 year to Gregorian year
    # Handle 月 (month conversion)
    elif '月' in value:
        month_map = {
            '1月': 'Jan.', '2月': 'Feb.', '3月': 'Mar.', '4月': 'Apr.',
            '5月': 'May', '6月': 'Jun.', '7月': 'Jul.', '8月': 'Aug.',
            '9月': 'Sep.', '10月': 'Oct.', '11月': 'Nov.', '12月': 'Dec.'
        }
        return month_map.get(value, value)  # Map months to their English abbreviations
    # Default to the original value if no match
    return value

# Ensure df is a copy
df = df.copy()

# Apply the transformation to the first column
df.loc[:, 'Month/Year'] = df['Month/Year'].apply(transform_date)

df
```

```python
# Ensure that month/year is a string to handle both months and years
df['Month/Year'] = df['Month/Year'].astype(str)

# Filter rows starting from '2023'
start_index = df[df['Month/Year'].str.contains('2023', na=False)].index[0]

# Filter the DataFrame from the start_index onward
df_filtered = df.loc[start_index:].reset_index(drop=True)

# Display the filtered DataFrame
df_filtered
```

```python
df_filtered = df_filtered[~df_filtered.iloc[:, 0].astype(str).str.match(r'\d{4}')]

# Reset the index after filtering
df_filtered.reset_index(drop=True, inplace=True)

# Create the 'year' column safely using .loc
df_filtered.loc[:, 'year'] = (df_filtered.index // 12) + 2023  # Start from 2023 for the first 12 rows

# Display the updated dataframe
df_filtered
```

```python
# Ensure the numerical columns are floats and then round them to 2 decimal places
df_filtered[['Total', 'Export', 'Import', 'Balance']] = df_filtered[['Total', 'Export', 'Import', 'Balance']].astype(float).round(2)

# Display the updated dataframe
df_filtered
```

```python
df_filtered['Month/Year'] = df_filtered['Month/Year'] + ' ' + df_filtered['year'].astype(str)

# Drop the 'year' column as it's no longer needed
df_filtered.drop(columns=['year'], inplace=True)

# Display the updated dataframe
df_filtered
```

```python
df_filtered.to_csv("taiwan-us-trade.csv", index=False)
```
